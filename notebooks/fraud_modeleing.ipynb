```python
# import Libraries
import pandas as pd
import numpy as np 
import sys
sys.path.append("..")
```

##### Load The Data Using Pandas

```python
x_train=np.load("../data/processed/X_train.npy",allow_pickle=True)
x_test=np.load("../data/processed/X_test.npy",allow_pickle=True)
y_train=np.load("../data/processed/y_train.npy",allow_pickle=True)
y_test=np.load("../data/processed/y_test.npy",allow_pickle=True)

```

```python
type(x_train)
```

```python
#import and create object of custom module
from src.model_trainer import FraudModelTrainer
trainer=FraudModelTrainer()
```

#### Train Baseline Model (Logistic Regression) with Custom Module

```python
lr=trainer.train_logistic_regression(x_train,y_train)
```

#### Evaluate The Logistic Regression Model

```python
lr_metrics=trainer.evaluate(lr,x_test,y_test,"Logistic Regression")
lr_metrics
```

#### Train Ensemble Model(Random Forest)

```python
rf=trainer.train_random_forest(x_train,y_train)
```

#### Evaluate The Random Forest Model

```python
rf_metrics=trainer.evaluate(rf,x_test,y_test,"Random Forest")
rf_metrics
```

#### Tune The Random Forest Model

```python
tune_rf=trainer.tune_random_forest(x_train,y_train)
tune_rf
```

```python
# get the model results
results =trainer.get_result_table()
results
```
